In this notebook, a dataset of millions of movies (including shows, etc) was downloaded ranging from the present back to the late 1800s. While several different looks and approaches were entertained, such as type of title, genre, and runtime, upon further inspection, the most interesting angle was simply volume of production as a function of time. Specifically, there has been an exponential increase in the number of productions per year, with the most dramatic increase happening after the year 2000. After some data cleaning and filtering out non-movies, the above histogram for number of movies per year simply and stunningly displays the staggering trajectory of film production.
